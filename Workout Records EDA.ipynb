```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('data.csv')
```

```python
x = df["Calories"].mean()
df["Calories"] = df["Calories"].fillna(x)
```

```python
df['Date']=pd.to_datetime(df['Date'], format='mixed')
```

```python
df.info()
```

```python
# Plotting Calories Over Time 
plt.figure(figsize=(15, 6))
sns.lineplot(x='Date', y='Calories', data=df, marker='o')
plt.title('Calories Burned Over Time')
```

```python
# Average calories by duration 
avg_calories_by_duration = df.groupby('Duration')['Calories'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 6))
barplot= sns.barplot(x='Duration', y='Calories', data=avg_calories_by_duration, hue= 'Duration', palette='viridis')
for index, row in avg_calories_by_duration.iterrows():
    barplot.text(row.name, row['Calories'] + 5, round(row['Calories'], 1), color='black', ha="center")
plt.title('Average Calories Burned by Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Average Calories')
```

```python
# Scatter plot between Calories and Pulse
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Pulse', y='Calories', data=df, hue='Duration', palette='muted')
plt.title('Calories vs Pulse')
```

```python
# Count of workouts by intensity 
def categorize_intensity(pulse):
    if pulse < 100:
        return 'Low Intensity'
    elif 100 <= pulse <= 120:
        return 'Moderate Intensity'
    else:
        return 'High Intensity'
df['Intensity'] = df['Pulse'].apply(categorize_intensity)
workout_count_by_intensity = df['Intensity'].value_counts().reset_index()
workout_count_by_intensity.columns = ['Intensity', 'Workout Count']
```

```python
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='Intensity', y='Workout Count', data=workout_count_by_intensity, hue='Intensity', palette='coolwarm')
for index, row in workout_count_by_intensity.iterrows():
    barplot.text(row.name, row['Workout Count'] + 0.5, row['Workout Count'], color='black', ha="center")
plt.title('Count of Workouts by Intensity')
plt.xlabel('Intensity')
plt.ylabel('Workout Count')
```
